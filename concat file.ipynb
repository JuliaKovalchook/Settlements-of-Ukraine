```python
import glob
import os
import pandas as pd
```

```python
path  = 'C:/Users/Administrator/Desktop/Новая папка/Одеса'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    region=filename.split("\\")[-1].split("raw_data_vilige_")[-1].split(".csv")[0]
    df['region']=region
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame
```

```python
frame.to_csv(r'C:/Users/Administrator/Desktop/Новая папка/raw_data_vilige_odesa.csv', index = False, header = True)
```
